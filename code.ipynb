```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
```

```python
raw_data = pd.read_csv('data.csv')
raw_data
```

## Data wrangling

```python
data_df = raw_data[raw_data['S.No'] != 'Total']
data_df
```

```python
tapioca = data_df[data_df['Name of the crop'] == 'Tapioca']
onion = data_df[data_df['Name of the crop'] == 'Onion']
chillies = data_df[data_df['Name of the crop'] == 'Chillies']
```

```python
tapioca_data = tapioca.to_numpy()
onion_data = onion.to_numpy()
chillies_data = chillies.to_numpy()

plt.figure(figsize=(10,5))
plt.scatter(tapioca_data[:,4], tapioca_data[:,6])
plt.title('Tapioca estimated production trends w.r.t area allocated', fontsize=20)
plt.xlabel('Area(in Hectare)', fontsize=15)
plt.ylabel('Estimated Production (Tonnes)', fontsize=15)
plt.show()
```

```python
# box_data = np.array([tapioca_data[:,5], onion_data[:,5], chillies_data[:,5] ]).T
box_data = np.array([tapioca_data[:,5]]).T
plt.figure(figsize=(10,5))
plt.boxplot(box_data)
plt.title('Estimated Average Yield per (Kg/Hectare) across districts', fontsize=20)
plt.xlabel('Tapioca', fontsize=15)
plt.ylabel('Estimated Average Yield per (Kg/Hectare)', fontsize=15)
# plt.xticks(('Tapioca', 'Onion', 'Chillies'))
plt.show()
```

```python
total_df = raw_data[raw_data['S.No'] == 'Total']
total_df
```

```python
total_data = total_df.to_numpy()
plt.figure(figsize=(10,5))
plt.bar(total_data[:,2], total_data[:,4])
plt.title('Total Area(in Hectare) for different crops', fontsize=20)
plt.xlabel('Crops', fontsize=15)
plt.ylabel('Area(in Hectare)', fontsize=15)
plt.show()
```
